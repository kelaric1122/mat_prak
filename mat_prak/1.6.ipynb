```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats
import seaborn as sns
import random
from math import fabs

def func(n,tetta):

    m1 = n*tetta/(n+1)
    d1 = n*tetta/((n+2)*((n+1)**2))
    norm1 = np.random.normal(loc = m1, scale = np.sqrt(d1), size = n )

    data_Xn = [np.max(np.random.uniform(0, tetta, size=i + 1)) for i in range(n)]
    data_Xn = np.array(data_Xn).flatten()
    plt.figure()
    plt.hist(data_Xn, bins = n//10 , label='Xn', alpha=0.5)
    plt.hist(norm1, bins = 5, label='Normal', alpha=0.5, color='red')

    plt.title('n = str(n)')
    plt.legend()
    plt.show()

tetta=float(input('введите tetta ') )
func(50,tetta)
func(100,tetta)
func(500,tetta)


#пункт б
def func_1( n, tetta):
      data= np.random.uniform(0,tetta , size= n)
      new_data =  [ (i + 1) *(tetta - np.max(np.random.uniform (0, tetta, size = i+1 ) ) ) for i in range(n)]
      plt.figure()
      x = np.arange(0,100,1)
      plt.plot(x, 175*stats.expon.pdf(x, scale = 1), label = 'exp(1)')
      sns.histplot(new_data, label = 'uniform', alpha = 0.5, color = 'red')
      plt.title('n = 1000')
      plt.legend()
func_1( 50, tetta)
func_1( 100, tetta)
func_1( 500, tetta)
#пункт d
n = 100
data1 = []
data2 = []
k = 0

for i in range(1000):
  X = np.random.uniform(low = 0, high = tetta, size = n)
  if ((fabs((n + 1)/n)*np.max([X]) - tetta) > fabs(2*np.mean(X) - tetta)):
    k = k + 1
  data1.append(((n + 1)/n)*np.max([X]))
  data2.append(2*np.mean(X))

if (k < 1000 - k):
  print('Оценка максимума ближе к тетте')
  print('Доля: ', 1000 - k)
else:
  print('Оценка среднего ближе к тетте')
  print('Доля: ', k)

plt.figure()
plt.hist(data1, alpha = 0.5, color = 'red', label = 'максимум')
plt.hist(data2, alpha = 0.5, color = 'blue', label = 'Среднее')
plt.axvline (x=tetta, color='green', linestyle='--')
plt.legend()
```

```python
## Задание 2б
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats
import seaborn as sns
import random
from math import fabs
n = int(input('Введите n '))
tetta = random.uniform(0, 5)
data1 = []
new_data1 = []
data2 = []

for i in range(n):

  cauchy_rv = stats.cauchy(loc = tetta)
  data1 = cauchy_rv.rvs(i+1)
  data1.sort()
  if (i+1 % 2) == 0:
    med = (data1[i+1//2] + data1[(i+1 - 1)//2])/2
  else:
    med = data1[(i+1 - 1)//2]
  new_data1.append(np.sqrt(i)*(med - tetta))

data2 = np.random.normal (loc = 0, scale = np.pi/2, size=n)

plt.figure()
plt.hist(new_data1, bins = 7, label = 'cauchy', alpha = 0.5, color = 'red')
plt.hist(data2, bins = 7,label = 'normal', alpha = 0.5, color = 'blue')
plt.title('n = 1000')
plt.legend()
```

```python
## Задача 5
from numpy.random.mtrand import laplace
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats
import seaborn as sns
import random
import math
from math import fabs
from math import exp

tetta = 3
n = int(input('Введите n '))
size = int(input('Введите size '))
b = float(input('Коэф масштаба '))
#b = float(input('Коэф сдвига '))

k = 0
new_data1 = []
new_data2 = []
new_data3 = []
new_data4 = []

for i in range(n):
  k = k+1
  laplace_rv = stats.laplace(loc = tetta, scale = b)
  data1 = laplace_rv.rvs(size)
  new_data2.append(np.sqrt(size)*(np.mean(data1) - tetta))
  data1.sort()
  X1 = stats.trim_mean(data1, 0.1)
  new_data3.append(np.sqrt(size)*(X1 - tetta))
  X2 = stats.trim_mean(data1, 0.3)
  new_data4.append(np.sqrt(size)*(X2 - tetta))
  if (size % 2) == 0:
    med = (data1[size//2] + data1[(size - 1)//2])/2
  else:
    med = data1[(size - 1)//2]
  new_data1.append(np.sqrt(size)*(med - tetta))

plt.figure()
plt.hist(new_data1, label = 'med', alpha = 0.5, color = 'red', bins = 20)
plt.hist(new_data2, label = 'Среднее', alpha = 0.5, color = 'green', bins = 20)
plt.hist(new_data3, label = 'Усеч. сред. a = 0.1', alpha = 0.5, color = 'blue', bins = 20)
plt.hist(new_data4, label = 'Усеч. сред. a = 0.3', alpha = 0.5, color = 'orange', bins = 20)
plt.legend()
```
