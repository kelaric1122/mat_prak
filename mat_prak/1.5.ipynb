```python
import matplotlib.pyplot as plt
import seaborn as sb
import numpy as np
import scipy.stats as stats
from scipy.stats import bernoulli as bern
from scipy.stats import expon as exp

n = int(input())

a = float(input())
b = float(input())
data = []
for i in range(n):
  bern_rv = stats.bernoulli(0.5)
  y = bern_rv.rvs(1)
  exp_rv = stats.expon(loc = a, scale = b)
  x = exp_rv.rvs(1)
  if y[0] == 0:
    data.append(x[0])
  else:
    t = (2 * a - x[0])
    data.append(t)

print(data)

plt.figure()
plt.hist(data, bins = 20, density = True)
plt.title('Лаплас')

data = []
for i in range(n):
  y = bern(1/2).rvs(size = 1)
  if y[0] == 0:
    x = exp(scale = 1).rvs(size = 1)
    data.append(x[0])
  elif y[0] == 1:
    x = exp(scale = 0.5).rvs(size = 1)
    t = (- x[0])
    data.append(t)

plt.figure()
plt.hist(data, bins = 20, density = True)
```

```python
from ast import AnnAssign
import math
import pandas as pd
import seaborn as sb
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import binom
from scipy.stats import norm
from pandas.plotting import parallel_coordinates
import statsmodels.api as sm
import scipy.stats as stats


n=1000
k=0
x=np.zeros(n)
y=np.zeros(n)
i=0

while k<n:
  X=np.random.uniform(-1, 1, 1)
  Y=np.random.uniform(0, 1, 1)
  while Y+np.fabs(X)>1:
    X=np.random.uniform(-1, 1, 1)
    Y=np.random.uniform(0, 1, 1)
  x[i]=X
  y[i]=Y
  k+=1
  i+=1



for i in range(n):
  plt.plot(x[i],y[i], 'b.', alpha=0.8)

u=np.linspace(-1, 1, 100)
v=[-np.fabs(t)+1 for t in u]
d=np.zeros(100)
plt.plot(u,v)
plt.plot(u,d)

plt.legend()
plt.show()
```
