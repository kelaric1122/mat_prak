```python
X = np.random.binomial(n=100, p=0.04, size=1000)

Y = np.random.poisson(lam=4, size=1000)

plt.figure()
plt.subplot(121)
plt.hist(X, bins=range(31), density=True)
plt.xlabel('X')
plt.title('X ~ Bin(100, 0.04)')
plt.subplot(122)
plt.hist(Y, bins=range(11), density=True)
plt.xlabel('Y')
plt.title('Y ~ Poiss(4)')

plt.tight_layout()

plt.figure()
plt.hist(X, bins=range(31), alpha=0.5, density=True, label='X ~ Bin(100, 0.04)')
plt.hist(Y, bins=range(11), alpha=0.5, density=True, label='Y ~ Poiss(4)')
plt.title('Comparison of X and Y')
plt.legend()
```

```python
from pandas.plotting import parallel_coordinates

titanic = sns.load_dataset('titanic')
titanic['who_is'] = titanic.apply(lambda x: 1 if x['who'] == 'man'  else 20, axis=1)

del titanic['who']
del titanic['sex']
del titanic['embarked']
del titanic['class']
del titanic['adult_male']
del titanic['deck']
del titanic['embark_town']
del titanic['alive']
del titanic['alone']
del titanic['parch']
del titanic['sibsp']
del titanic['fare']

data = titanic.sample(n = 30)

plt.figure(figsize=(6, 5))
parallel_coordinates(data, 'survived', colormap='gist_rainbow')
```
